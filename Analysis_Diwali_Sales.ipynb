# Exploratory Data Analysis

# Gender

```python
df.columns
```

```python
# Plot a coubntplot graph in Gender Comparing either male or femal purchase more
g_plot = sns.countplot(x = 'Gender', data = df)
# if we need a count in bars
for bars in g_plot.containers:
    g_plot.bar_label(bars)
```

```python
sale_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sale = sns.barplot(x = 'Gender', y= 'Amount', data= sale_gen)
for bars in sale.containers:
    sale.bar_label(bars)
```

- From above Graph we can see that most of the buyers are Females even the purchasing power are Greater than male

# State

```python
sate_order = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)
```

```python
#total order from top state
sns.set(rc={'figure.figsize':(25, 10)})
ax = sns.barplot(x = 'State',y = 'Orders', data = sate_order)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total sale/amount from top state
state_sale = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
ax = sns.barplot(x = 'State', y = 'Amount', data = state_sale)
for bars in ax.containers:
    ax.bar_label(bars)
```

- From the above graph we can see that the most of the orders are from Uttar Pardesh, Maharashtra and Kernataka respectively 

# Martial Status

```python
# Replace values in 'Marital_Status' column
df['Marital_Status'] = df['Marital_Status'].replace({0: 'Married', 1: 'Unmarried'})
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status', hue = 'Gender' )
sns.set(rc={'figure.figsize':(7, 5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Replace values in 'Marital_Status' column
df['Marital_Status'] = df['Marital_Status'].replace({0: 'Married', 1: 'Unmarried'})

# Plot the count plot with updated labels
ax = sns.countplot(data=df, x='Marital_Status', hue='Gender')
sns.set(rc={'figure.figsize':(10, 10)})
# Add labels to the bars
for bars in ax.containers:
    ax.bar_label(bars, label_type='center')

# Set the text labels inside each bar
for p in ax.patches:
    ax.annotate(f"{p.get_height()}", (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center',
                xytext=(0, 5), textcoords='offset points')

# Show the plot
plt.show()
```

```python
mar_sale = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sns.set(rc={'figure.figsize':(7, 5)})
ax = sns.barplot(x= 'Marital_Status', y= 'Amount', data= mar_sale, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

- From the above graph we can see that most of the buyers are married women and theay have high purchasing power
